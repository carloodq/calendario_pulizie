```python
import pandas as pd
from datetime import datetime
df = pd.read_csv("turni.csv")
```

```python
df.head()
```

```python
from datetime import timedelta


def get_data_inizio(row):
    base_time = datetime.strptime(row['periodo'].split('-')[0], "%d/%m/%Y")
    if row['turno'] == 'mattino':
        return base_time + timedelta(hours=6)
    elif row['turno'] == 'pomeriggio':
        return base_time + timedelta(hours=12)
    elif row['turno'] == 'sera':
        return base_time + timedelta(hours=18)
    else:
        return base_time  # default if turno is missing

df['data_inizio'] = df.apply(get_data_inizio, axis=1)
df['data_fine'] = df['data_inizio']+ timedelta(hours=6)
```
